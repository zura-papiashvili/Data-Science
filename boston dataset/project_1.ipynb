```python
import lineare_regression as lr
```

```python
# IMPORT BOSTON DATASET
from sklearn.datasets import load_boston
dataset = load_boston()
```

```python
# 1. Split dataset into train and test
#DIVISION FACTOR 80%
thr=int(len(dataset.target)*0.8)

y_train=dataset.target[:thr]
y_test=dataset.target[thr:]
X_train=dataset.data[:thr]
X_test=dataset.data[thr:]
```

```python
# 2. Feateure exploring:
#Histogram of target values(prices):
plt.hist(y_train,bins=50)
plt.title("Prices")
plt.show()
```

```python
#Histogram of room numbers:

plt.hist(X_train[:,5], bins=50)
plt.title("room numbers")
plt.show()
```

```python
#Scatter plot of room numbers vs price

plt.scatter(y_train,X_train[:,5])
plt.title("room number vs price")
plt.xlabel("price")
plt.ylabel("room numbers")
plt.show()
```

```python
#Scatter LSTAT vs price

plt.scatter(y_train,X_train[:,-1])
plt.title("% lower status of the population vs price")
plt.xlabel("price")
plt.ylabel("LSTAT")
plt.show()
```

```python
# 3. Make a linear regression model of the training data
theta=lr.LR_fit(X_train,y_train)
print("theta: \n",theta)
```

```python
# 4. test and evaluation
# make prediction for testing dataset
predict=lr.LR_predict(X_test,theta)
predict
```

```python
r_score=lr.r2_score(X_test,y_test,theta)
print("r_score: ",r_score)
```

```python
plt.figure(figsize=(10,10))
plt.scatter(predict,y_test)
plt.xlabel("predicted values")
plt.ylabel("actual value")
plt.show()
```

```python
import numpy as np
mse = ((np.square(y_test - predict)).mean(axis=0))
print("mse:",mse)
```
