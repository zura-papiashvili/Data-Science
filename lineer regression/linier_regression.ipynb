```python
import lineare_regression as lr
import numpy as np
```

```python
X=np.array([[0],[6],[12]])
y=np.array([[13.5],[10.5],[3.0]])
```

```python
lr.extend_matrix(X)
```

```python
theta=lr.LR_fit(X,y)
print("theta",theta)
```

```python
lr.LR_predict(X,theta)
```

```python
lr.r2_score(X,y,theta)
```

```python
h_theta = lr.LR_predict(X,theta)
y_mean = np.mean(y)
c = np.sum((y-h_theta)**2)
d = np.sum((y-y_mean)**2)
r2 = 1-(c/d)
r2
```
